# Mapa de Calor

```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sb.heatmap(df.corr(), linewidths=0.1, vmax=1.0, square=True, 
           cmap=colormap, linecolor='white', annot=True);
```

# Histogramas


*Cantidad de Pokemon por tipo principal (tipo 1)*

```python
df["Type 1"].value_counts().plot(kind='bar')
```

*Cantidad de Pokemon por tipo secundario (tipo 2)*

```python
df["Type 2"].value_counts().plot(kind='bar')
```

*Cantidad de Pokemon por cada tipo (tipo 1 + tipo 2)*

```python
type1Count = df['Type 1'].value_counts()
type2Count= df['Type 2'].value_counts()
typeTotal = type1Count + type2Count
typeTotal.plot(kind='bar')
```

*NÃºmero de Pokemon por cada generaciÃ³n*

```python
df['Generation'].value_counts().plot.pie();
```

*Defensa con respecto al ataque*

```python
plt.plot(df.Attack, df.Defense, '*')
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.show()
```
